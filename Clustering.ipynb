1. What is unsupervised learning in the context of machine learning?
- Unsupervised learning is a type of machine learning where the model is trained on data without labeled responses. The goal is to discover hidden patterns, groupings, or structures in the data. Clustering is a common unsupervised technique, used to categorize data into meaningful groups (clusters) based on similarity. It's helpful for exploratory data analysis, market segmentation, anomaly detection, and more.

2. How does K-Means clustering algorithm work?
- K-Means partitions data into K clusters by minimizing the distance between data points and their assigned cluster centroid. The algorithm starts by initializing K centroids, then iteratively assigns points to the nearest centroid and recalculates centroids based on the mean of assigned points. This process continues until centroids stabilize or a stopping condition is met. It 's simple, efficient, and works best with spherical clusters.



3. Explain the concept of a dendrogram in hierarchical clustering.
- A dendrogram is a tree-like diagram used to illustrate the arrangement of clusters formed by hierarchical clustering. It shows how individual elements merge into clusters step-by-step. The vertical axis represents the distance or dissimilarity between clusters. By cutting the dendrogram at a specific level, users can decide the number of clusters. It provides insight into the structure and relationship among clusters.

4. What is the main difference between K-Means and Hierarchical Clustering?
- K-Means is a partitional clustering method that requires the number of clusters to be defined in advance, and it partitions the data into non-overlapping subsets. Hierarchical clustering builds a tree of clusters without needing the number of clusters upfront. It can be agglomerative (bottom-up) or divisive (top-down). K-Means is faster but less informative, while hierarchical is more descriptive but computationally expensive.



5. What are the advantages of DBSCAN over K-Means?
- DBSCAN does not require the number of clusters beforehand and can discover clusters of arbitrary shapes, unlike K-Means, which works best with spherical clusters. DBSCAN can also identify and handle noise or outliers effectively, which K-Means struggles with. Moreover, DBSCAN can handle varying cluster densities, while K-Means may perform poorly when cluster sizes and densities differ significantly.



6. When would you use Silhouette Score in clustering?
- Silhouette Score is used to evaluate the quality of clustering. It measures how similar a point is to its own cluster compared to other clusters. The score ranges from -1 to 1; a high value indicates well-separated clusters. It is especially useful when the true labels are unknown and helps in determining the optimal number of clusters by comparing clustering performance across K values.



7. What are the limitations of Hierarchical Clustering?
- Hierarchical clustering is computationally expensive (O(n²) or higher), making it unsuitable for large datasets. It also lacks robustness—once a merge or split is made, it cannot be undone. Furthermore, it doesn't handle noise or outliers well, and results can be sensitive to the chosen linkage method and distance metric. It also doesn't scale efficiently with increasing data dimensions.

8. Why is feature scaling important in clustering algorithms like K-Means?
- Clustering algorithms like K-Means rely on distance measures (usually Euclidean distance), so features with larger ranges can dominate the clustering process. Feature scaling standardizes the data, ensuring each feature contributes equally. Without scaling, results may be biased toward features with larger values, leading to incorrect or suboptimal clusters. Techniques like Min-Max scaling or Standardization (Z-score normalization) are commonly used.



9. How does DBSCAN identify noise points?
- DBSCAN classifies points based on density. If a point has fewer than the minimum required neighbors within the specified radius (epsilon), it is labeled as noise or an outlier. These points do not belong to any cluster. Unlike K-Means, DBSCAN doesn't force every point into a cluster, making it effective for identifying sparse regions and anomalies in data.

10. Define inertia in the context of K-Means.
- Inertia is the sum of squared distances between data points and their respective cluster centroids. It represents how tightly grouped the clusters are—the lower the inertia, the more compact the clusters. Inertia is used as a metric to evaluate and compare clustering results, especially when determining the optimal number of clusters using the elbow method in K-Means.

11. What is the elbow method in K-Means clustering?
- The elbow method helps determine the optimal number of clusters (K) in K-Means by plotting the inertia (within-cluster sum of squares) against various K values. As K increases, inertia decreases. The point at which the decrease in inertia slows down sharply (forming an “elbow” shape) indicates the optimal K. Beyond this point, adding more clusters gives diminishing returns.



12. Describe the concept of "density" in DBSCAN.
- In DBSCAN, density refers to the number of data points within a specified radius (epsilon) of a point. If this number exceeds a threshold (min_samples), the point is considered a core point and part of a dense region. Points within the radius of a core point are assigned to the same cluster. Density helps DBSCAN find clusters of arbitrary shapes and detect outliers.



13. Can hierarchical clustering be used on categorical data?
- Yes, but with modifications. Standard hierarchical clustering requires a distance metric, which is not directly applicable to categorical data. To use it, categorical features must be transformed using techniques like one-hot encoding or converted into a similarity/dissimilarity matrix (e.g., using Jaccard distance). Specialized algorithms like Gower distance can also handle mixed data types for hierarchical clustering.



14. What does a negative Silhouette Score indicate?
- A negative Silhouette Score indicates that a data point is likely assigned to the wrong cluster. It means the point is closer to a neighboring cluster than to the one it's currently assigned to. This suggests poor clustering and may indicate overlapping clusters, an incorrect number of clusters, or ineffective clustering features. Negative values are a sign to re-evaluate the clustering setup.



15. Explain the term "linkage criteria" in hierarchical clustering.
- Linkage criteria determine how distances between clusters are computed during hierarchical clustering. Common types include:
- Single linkage: minimum distance between any two points
- Complete linkage: maximum distance
- Average linkage: average of all pairwise distances
- Ward's linkage: minimizes total within-cluster variance
Each method affects the shape and structure of the resulting dendrogram and influences clustering results.



16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
- K-Means assumes clusters are spherical and of equal size/density. When cluster sizes or densities vary, K-Means tends to assign points incorrectly—smaller or denser clusters may get merged, and larger ones may get split. It also minimizes distance from centroids, ignoring actual density distribution. Algorithms like DBSCAN or Gaussian Mixture Models are better suited for such data

17. What are the core parameters in DBSCAN, and how do they influence clustering?
- The two main parameters are:
- epsilon (ε): defines the radius for neighborhood search.
min_samples: minimum number of points required to form a dense region.
These control how clusters are formed. A small ε may cause too many small clusters or noise, while a large ε may merge distinct clusters. min_samples influences cluster density sensitivity. Proper tuning is essential for accurate clustering.

18. How does K-Means++ improve upon standard K-Means initialization?
- K-Means++ improves the initial selection of centroids to avoid poor clustering due to random initialization. It selects the first centroid randomly, then picks subsequent centroids from the remaining data points based on a probability proportional to their distance from existing centroids. This results in better spread-out starting points, faster convergence, and improved clustering quality compared to standard K-Means.



19. What is agglomerative clustering?
- Agglomerative clustering is a bottom-up hierarchical clustering method. It starts with each data point as its own cluster, then iteratively merges the two closest clusters based on a linkage criterion, continuing until all points form a single cluster or a stopping condition is met. The process forms a dendrogram and is commonly used to explore data structure.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?
- While inertia measures how compact clusters are, it doesn't account for separation between clusters. Silhouette Score considers both cohesion (intra-cluster distance) and separation (inter-cluster distance), giving a more balanced evaluation. It ranges from -1 to 1, with higher values indicating well-formed clusters. Thus, it offers deeper insight into clustering quality than inertia alone, especially when comparing different K values.

```python
#21.Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot?
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 2: Generate Synthetic Data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Step 3: Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Step 4: Visualize the Clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')
plt.title('K-Means Clustering with 4 Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

```

```python
#23.Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot?
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Step 2: Generate Synthetic Moon-Shaped Data
X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

# Step 3: Apply DBSCAN Clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Step 4: Plot Clusters and Outliers
plt.figure(figsize=(8, 6))

# Unique labels including -1 for noise
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for label, col in zip(unique_labels, colors):
    if label == -1:
        # Black used for noise (outliers)
        col = [0, 0, 0, 1]
        label_name = 'Outliers'
    else:
        label_name = f'Cluster {label}'

    class_member_mask = (labels == label)
    xy = X[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], s=50, c=[col], label=label_name)

plt.title('DBSCAN on make_moons data (Outliers Highlighted)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

```

```python
#29.Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart?
# Step 1: Import Libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 2: Generate Synthetic Data
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Step 3: Evaluate Silhouette Scores for k = 2 to 5
silhouette_scores = []

k_values = range(2, 6)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Step 4: Plot Silhouette Scores as Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for K = 2 to 5")
plt.xticks(k_values)
plt.grid(True, axis='y')
plt.show()
```

```python
#31.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries ?
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# Step 2: Generate Overlapping Data
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=2.0, random_state=42)

# Step 3: Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)
centers = kmeans.cluster_centers_

# Step 4: Create a Mesh Grid for Decision Boundaries
h = 0.05  # step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Step 5: Plot Decision Boundaries and Clusters
plt.figure(figsize=(10, 6))
cmap_background = ListedColormap(['#FFCCCC', '#CCE5FF', '#CCFFCC'])
cmap_clusters = ListedColormap(['red', 'blue', 'green'])

# Plot decision boundary
plt.contourf(xx, yy, Z, cmap=cmap_background, alpha=0.5)

# Plot data points
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap=cmap_clusters, edgecolor='k', s=50)

# Plot centroids
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, marker='X', label='Centroids')

plt.title("K-Means Clustering with Decision Boundaries (Overlapping Clusters)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()

```

```python
#44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering ?
# Step 1: Import Libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np

# Step 2: Load Iris Dataset
iris = load_iris()
X = iris.data

# Step 3: Apply KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(X)

# Step 4: Compute Silhouette Scores for Each Sample
silhouette_vals = silhouette_samples(X, cluster_labels)

# Step 5: Plot Silhouette Scores
plt.figure(figsize=(10, 6))
y_lower = 10
for i in range(3):
    ith_cluster_silhouette_vals = silhouette_vals[cluster_labels == i]
    ith_cluster_silhouette_vals.sort()

    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_vals,
                      alpha=0.7, label=f'Cluster {i}')

    y_lower = y_upper + 10  # Space between clusters

# Plot the average silhouette score as a vertical line
avg_score = silhouette_score(X, cluster_labels)
plt.axvline(x=avg_score, color="red", linestyle="--", label=f"Avg Score: {avg_score:.2f}")

plt.title("Silhouette Plot for Each Sample (Iris Dataset, KMeans)")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Sample Index by Cluster")
plt.legend()
plt.grid(True)
plt.show()
```
